```python
import pandas as pd
```

```python
data = {
    'Name': ['Ali', 'Sara', 'Ahmed', 'Ayesha'],
    'Age': [20, 22, 19, 21],
    'Marks1': [80, 85, 78, 90],
    'Marks2': [70, 88, 82, 95],
    'CGPA': [3.2, 3.8, 2.9, 3.9],
    'Date': ['2024-01-10', '2024-02-15', '2024-03-20', '2024-04-25']
}

df = pd.DataFrame(data)
```

```python
df
```

Arithmetic Operations

```python
df['Total'] = df['Marks1'] + df['Marks2']
df
```

Statistical Operations

```python
df['CGPA'].mean()
```

Logical Operations

```python
df['Passed'] = df['CGPA'] >= 3.0
df
```

String Operations

```python
df['Name_Upper'] = df['Name'].str.upper()
df
```

Sorting Operations

```python
df.sort_values('CGPA')
```

```python
df.sort_values('Age', ascending=False)
```

DateTime Operations

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df['Year'] = df['Date'].dt.year
df
```
