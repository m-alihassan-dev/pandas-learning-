```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

Finding Missing Data

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
df.isna().sum().sum()
```

Removing Missing Data

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=3)
```

Filling Missing Data

```python
df
```

```python
df.fillna(0)
```

```python
df['A'].fillna(df['A'].mean())
```

```python
df.fillna(df.mean())
```

```python
df.fillna(df.std())
```

```python
df.fillna(df.ffill())
```

```python
df.fillna(df.bfill())
```

```python
values = {'A': 10, 'B': 0, 'C': 50, 'D': 100}
df.fillna(value=values)
```
