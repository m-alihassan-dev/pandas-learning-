Importing Libraries

```python
import numpy as np
import pandas as pd
from datetime import datetime
import re
```

Loading Dataset

```python
df = pd.read_csv('anime.csv')
df.head()
```

Extract Anime Type

```python
def extract_type(title):
    if 'Movie' in title:
        return 'Movie'
    elif 'OVA' in title:
        return 'OVA'
    elif 'Special' in title:
        return 'Special'
    else:
        return 'TV'
    
df['Type'] = df['Title'].apply(extract_type)
df.head()
```

Extracing Episodes

```python
def extract_episodes(title):
    match = re.search(r'\((\d+)', title)
    return int(match.group(1)) if match else np.nan

df['Episodes'] = df['Title'].apply(extract_episodes)
df.head()
```

Extracting Release Year

```python
def extract_year(title):
    match = re.search(r'(19|20)(\d{2})', title)
    return int(match.group()) if match else np.nan

df['Release Year'] = df['Title'].apply(extract_year)
df.head()
```

Creating Anime Length Feature (Short/Long)

```python
df['Length Type'] = df['Episodes'].apply(
    lambda x: 'Long' if x >= 50 else 'Short'
)

df.head()
```

Creating Top Tier Feature

```python
df['Top Tier'] = df['Score'].apply(
    lambda x: 'Yes' if x >= 8.5 else 'No'
)

df.head()
```

Creating Anime Age Feature

```python
current_year = datetime.now().year
df['Anime Age'] = current_year - df['Release Year']
df.head()
```
