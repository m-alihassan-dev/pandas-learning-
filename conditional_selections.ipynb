```python
import pandas as pd
```

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}

df = pd.DataFrame(data)
```

```python
df
```

Single Condition

```python
df[df['Age'] > 30]
```

Multiple Condition

```python
df[(df['Age'] > 30) & (df['Salary'] > 70000)]
```

Conditional Selection using loc (Recommended)

```python
df.loc[df['Age'] > 30, ['Name', 'Salary']]
```
