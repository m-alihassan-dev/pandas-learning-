```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['CS', 'IT', 'CS', 'IT', 'SE'],
    'Student': ['Ali', 'Sara', 'Ahmed', 'Usman', 'Ahsan'],
    'CGPA': [3.5, 3.8, 3.2, 3.9, 3.6],
    'Age': [21, 22, 20, 23, 21]
})
```

```python
df
```

Average CGPA per Department

```python
df.groupby('Department')['CGPA'].mean()
```

Sum of CGPA of per Department

```python
df.groupby('Department')['CGPA'].sum()
```

GroupBy + Multiple Aggregations

```python
df.groupby('Department')['CGPA'].agg(['mean', 'max', 'min'])
```

GroupBy + Multiple Columns + Multiple Aggregations

```python
df.groupby('Department').agg({
    'CGPA': ['mean', 'max'],
    'Age': 'mean'
})
```

Rename Aggregated Columns

```python
df.groupby('Department').agg(
    Avg_CGPA = ('CGPA', 'mean'),
    Max_CGPA = ('CGPA', 'max'),
    Min_CGPA = ('CGPA',  'min'),
    Avg_Age = ('Age', 'mean')
)
```

Size

```python
df.groupby('Department').size()
```

Filter Groups after Aggregation

```python
df.groupby('Department').filter(lambda x: x['CGPA'].mean() > 3.5)
```
