```python
import pandas as pd
```

Creating from List

```python
list = [10, 20, 30, 40]
s = pd.Series(list)
print(s)
```

```python
list = [10, 20, 30, 40]
s = pd.Series(list, index=['a', 'b', 'c', 'd'])
print(s)
```

Creating from Dictionary

```python
data = {'Math': 86, 'English': 90, 'Science': 95}
s = pd.Series(data)
print(s)
```
