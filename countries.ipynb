```python
import numpy as np
import pandas as pd
```

Loading Dataset

```python
df = pd.read_csv('countries.csv')
df.head()
```

Basic Information about Dataset

```python
df.info()
```

Dataset Shape (Rows & Columns)

```python
df.shape
```

Column Names

```python
df.columns
```

Sort Countries by Population (Top 5)

```python
df.sort_values('population', ascending=False, inplace=True)
df.head()
```

Total Population of all Countries Combined

```python
total_population = df['population'].sum()
print(f'Total population: {total_population}')
```

Select Specific Columns

```python
df[['country', 'population']].head(10)
```

List all Unique Components in Dataset

```python
continents = df['continent'].unique()
print(f'Continents: {continents}')
```

Country with Highest GDP

```python
highest_gdp_country = df.loc[df['gdp'].idxmax(), 'country']
print(f'Country with highest GDP: {highest_gdp_country}')
```

Countries present in Asia

```python
asia_countries = df[df['continent'] == 'Asia']
asia_countries.head()
```

Create a New Column called gdp_per_capita

```python
df['gdp_per_capita'] = df['gdp'] / df['population']
df[['country', 'gdp_per_capita']].head()
```

Country with Highest GDP per Capita

```python
richest_per_capita = df.loc[df['gdp_per_capita'].idxmax(), 'country']
print(f'Country with highest GDP per capita: {richest_per_capita}')
```

Country with Smallest Land Area

```python
smallest_country = df.loc[df['land_area'].idxmin(), 'country']
print(f'Smallest country by area: {smallest_country}')
```

Top 5 Countries with Highest Inflation Rates

```python
high_inflation = df.sort_values('inflation', ascending=False).head(5)
high_inflation[['country', 'inflation']]
```

Slice the Dataframe to show only the country, capital_city, and currency for the First 10 Rows.

```python
subset = df.loc[:9, ['country', 'population', 'gdp']]
print(subset)
```

Average Population by Region

```python
df.groupby('region')['population'].mean()
```

Country Count per Region

```python
df.groupby('region')['country'].count()
```
