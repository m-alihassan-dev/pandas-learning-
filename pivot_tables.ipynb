```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Department': ['IT', 'IT', 'HR', 'HR'],
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'Salary': [50000, 55000, 40000, 42000]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
pd.pivot_table(df, values='Salary', index='Department', columns='Gender', aggfunc='mean')
```

```python
pd.pivot_table(df, values='Salary', index='Department', aggfunc=['mean', 'sum'])
```

Cross Tabluation (crosstab)

```python
pd.crosstab(df['Department'], df['Gender'])
```
